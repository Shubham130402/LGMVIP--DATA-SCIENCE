```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img=cv2.imread("owl_image.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.title("Owl", fontsize= 20)
plt.imshow(img)
plt.show()
```

```python
img=cv2.imread("Team.png")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.title("C-Helix Farewell", fontsize= 20)
plt.imshow(img)
plt.show()
```

```python
gray_img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.title("Gray Color of Owl")
plt.imshow(gray_img,cmap="gray")
plt.show()
```

```python
gray_img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.title("Gray Color of Team")
plt.imshow(gray_img,cmap="gray")
plt.show()
```

```python
invert_img= 255-gray_img
plt.imshow(invert_img,cmap="gray")
plt.show()
```

```python
invert_img= 255-gray_img
plt.imshow(invert_img,cmap="gray")
plt.show()
```

```python
blur_img=cv2.GaussianBlur(invert_img,(21,21),0)
inverted_blur_img=255-blur_img
```

```python
pencil_sketch=cv2.divide(gray_img,inverted_blur_img,scale=256.0)
plt.title("Pencil Sketch of Team", fontsize=18)
plt.imshow(pencil_sketch,cmap="gray")
plt.show()
```
